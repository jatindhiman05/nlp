```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
nlp.pipe_names
```

```python
text = nlp("Tesla Inc is going to acquire twitter for $45 billion")

for ent in text.ents:
    print(ent.text, " | ", ent.label_," |", spacy.explain(ent.label_))
```

```python
from spacy import displacy

displacy.render(text, style="ent")
```

```python
nlp.pipe_labels
```

### Setting own entities

```python
doc = nlp("Tesla Inc is going to acquire twitter for forty five billion dollars")
for ent in doc.ents:
    print(ent.text, " |", ent.label_)   
```

```python
S = doc[3:6]
S
```

```python
from spacy. tokens import Span

s1 = Span(doc, 0, 2, label='ORG')
s2 = Span(doc, 8,11, label='ORG')

doc.set_ents([s1,s2], default='unmodified')
```

```python
for ent in doc.ents:
    print(ent.text, " | ", ent.label_)
```
